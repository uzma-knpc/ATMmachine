```python
#Google Gimini chatmodal install

from langchain_google_genai import ChatGoogleGenerativeAI


llm = ChatGoogleGenerativeAI(api_key=GOOGLE_API_KEY,model="gemini-1.5-flash",temperature=0.7)
```

```python
#llm.invoke("what is nuclear pnra")
```

```python
from langchain_pinecone import PineconeVectorStore
vectorstore = PineconeVectorStore(index_name=index_name, embedding=embeddings)
```

```python
#list(dir(vectorstore))
```

```python
vectorstore.similarity_search("DMP about elution")
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k":4})
```

```python
retriever.get_relevant_documents("what is exposure limit of employees")
```

```python
#combine Retriver and llm
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="refine",  # Other options: "map_reduce", "refine"
    retriever=retriever,
    return_source_documents=True,
    verbose=True,
)
```

```python
query= "what is CT stand for?"
result = qa_chain.invoke({"query": query})
result["result"]
print(f"query:{query}")
result = qa_chain.invoke({"query": query})
result["result"]
```
